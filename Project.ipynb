```python
#Using OOPS
class Hotel:
    def __init__(self):
        self.total = 0
        self.bill = []

    def take_order(self):
        try:
            n = int(input("How many items? "))
            for i in range(1, n + 1):
                name = input("Item name: ")
                qty = int(input("Quantity: "))
                rate = int(input("Rate: "))
                amt = qty * rate
                self.total += amt
                self.bill.append((i, name, qty, rate, amt))
        except ValueError:
            print("Invalid input")
        else:
            print("Order taken")

    def deliver_order(self):
        if self.bill:
            print("Order delivered")
        else:
            print("No order")

    def print_bill(self):
        print("=" * 50)
        print("{:^50}".format("Star Hotel"))
        print("=" * 50)
        print("{:<3} {:<15} {:<5} {:<7} {:<7}".format(
            "SR", "ITEM", "QTY", "RATE", "PRICE"
        ))
        print("=" * 50)
        for item in self.bill:
            print("{:<3} {:<15} {:<5} {:<7} {:<7}".format(
                item[0], item[1], item[2], item[3], item[4]
            ))
        print("=" * 50)
        print("{:<30} {}".format("TOTAL AMOUNT", self.total))
        print("=" * 50)


h = Hotel()
h.take_order()
h.deliver_order()
h.print_bill()
```

```python
# CONNECTING TO THE DATABASE 
import pyodbc
from datetime import datetime

class Hotel:
    def __init__(self):
        self.bill = []
        self.total = 0

        try:
            self.conn = pyodbc.connect(
                r"Driver={ODBC Driver 17 for SQL Server};"
                r"Server=DESKTOP-BSQKA85\SQLEXPRESS;"
                r"Database=Hotel_Bill;"
                r"Trusted_Connection=yes;"
            )
            self.cur = self.conn.cursor()
            print("Connected to Database")
        except Exception as e:
            print("Database Connection Failed:", e)

    def show_menu(self):
        print("\n------ MENU ------")
        self.cur.execute("SELECT itemname, price FROM Menu")
        for row in self.cur.fetchall():
            print(f"{row.itemname} - Rs.{row.price}")
        print("------------------")

    def take_order(self):
        try:
            n = int(input("\nHow many items do you want to order? "))

            for _ in range(n):
                item = input("Enter item name: ")

                self.cur.execute(
                    "SELECT price FROM Menu WHERE itemname = ?", item
                )
                row = self.cur.fetchone()

                if not row:
                    print("Item not found in menu")
                    continue

                price = row.price
                qty = int(input("Enter quantity: "))
                total_price = qty * price
                self.total += total_price

                self.bill.append((item, qty, price, total_price))

                
                self.cur.execute(
                    "INSERT INTO orders (itemname, qty, price, total) VALUES (?, ?, ?, ?)",
                    item, qty, price, total_price
                )

            self.conn.commit()
            print("Order saved successfully")

        except ValueError:
            print("Invalid input. Please enter numbers only.")

    def print_bill(self):
        print("\n" + "=" * 50)
        print("{:^50}".format("STAR HOTEL"))
        print("=" * 50)
        print("{:<15} {:<10} {:<10} {:<10}".format(
            "ITEM", "QTY", "PRICE", "TOTAL"
        ))
        print("-" * 50)

        for item in self.bill:
            print("{:<15} {:<10} {:<10} {:<10}".format(
                item[0], item[1], item[2], item[3]
            ))

        print("-" * 50)
        print("{:<35} Rs.{}".format("TOTAL AMOUNT:", self.total))
        print("=" * 50)



h = Hotel()
h.show_menu()
h.take_order()

choice = input("\nDo you want to print the bill? (yes/no): ").lower()

if choice == "yes":
    h.print_bill()
else:
    print("Bill not printed")
```
