```python
#Using Class
class RestaurantBill:
    def __init__(self):
        self.bill = []
        self.total = 0

    def take_order(self):
        n = int(input("How many items do you want to eat? "))

        for i in range(1, n + 1):
            item = {}
            item["sr"] = i
            item["menu"] = input("Enter the menu : ")
            item["quantity"] = int(input("Enter the quantity : "))
            item["price"] = int(input("Enter the price : "))

            item["amount"] = item["quantity"] * item["price"]
            self.total += item["amount"]

            self.bill.append(item)

    def print_bill(self):
        print("=" * 70)
        print("{:^70}".format("Welcome Hotel Star"))
        print("=" * 70)

        print("{:<5} {:<25} {:<10} {:<10}".format(
            "Sr", "Menu", "Quantity", "Price"
        ))
        print("-" * 70)

        for item in self.bill:
            print("{:<5} {:<25} {:<10} {:<10}".format(
                item["sr"],
                item["menu"],
                item["quantity"],
                item["price"]
            ))

        print("=" * 70)
        print("{:<40} {:<10}".format("Total", self.total))
        print("=" * 70)


h = RestaurantBill()
h.take_order()

choice = input("Do you want to print the bill? (yes/no): ")

if choice.lower() == "yes":
    h.print_bill()
else:
    print("Bill not printed")

```

```python
#Try Except Else
class RestaurantBill:
    def __init__(self):
        self.bill = []
        self.total = 0

    def take_order(self):
        try:
            n = int(input("How many items do you want to eat? "))

            for i in range(1, n + 1):
                item = {}
                item["sr"] = i
                item["menu"] = input("Enter the menu : ")
                item["quantity"] = int(input("Enter the quantity : "))
                item["price"] = int(input("Enter the price : "))

                item["amount"] = item["quantity"] * item["price"]
                self.total += item["amount"]

                self.bill.append(item)

        except ValueError:
            print("Please enter numbers only for quantity and price")

        else:
            print("Order taken successfully")

    def print_bill(self):
        print("=" * 70)
        print("{:^70}".format("Welcome Hotel Star"))
        print("=" * 70)

        print("{:<5} {:<25} {:<10} {:<10}".format(
            "Sr", "Menu", "Quantity", "Price"
        ))
        print("-" * 70)

        for item in self.bill:
            print("{:<5} {:<25} {:<10} {:<10}".format(
                item["sr"],
                item["menu"],
                item["quantity"],
                item["price"]
            ))

        print("=" * 70)
        print("{:<40} {:<10}".format("Total", self.total))
        print("=" * 70)


h = RestaurantBill()
h.take_order()

choice = input("Do you want to print the bill? (yes/no): ")

if choice.lower() == "yes":
    h.print_bill()
else:
    print("Bill not printed")
```

```python
#Using OOPS
class Hotel:
    def __init__(self):
        self.total = 0
        self.bill = []

    def take_order(self):
        try:
            n = int(input("How many items? "))
            for i in range(1, n + 1):
                name = input("Item name: ")
                qty = int(input("Quantity: "))
                rate = int(input("Rate: "))
                amt = qty * rate
                self.total += amt
                self.bill.append((i, name, qty, rate, amt))
        except ValueError:
            print("Invalid input")
        else:
            print("Order taken")

    def deliver_order(self):
        if self.bill:
            print("Order delivered")
        else:
            print("No order")

    def print_bill(self):
        print("=" * 50)
        print("{:^50}".format("Star Hotel"))
        print("=" * 50)
        print("{:<3} {:<15} {:<5} {:<7} {:<7}".format(
            "SR", "ITEM", "QTY", "RATE", "PRICE"
        ))
        print("=" * 50)
        for item in self.bill:
            print("{:<3} {:<15} {:<5} {:<7} {:<7}".format(
                item[0], item[1], item[2], item[3], item[4]
            ))
        print("=" * 50)
        print("{:<30} {}".format("TOTAL AMOUNT", self.total))
        print("=" * 50)


h = Hotel()
h.take_order()
h.deliver_order()
h.print_bill()
```

```python
#13. Ask user to opt for courses for master degree based on the following L1 = [“HR”, “Finance”, “Marketing”, “DS”]
#Based on above subject there are two different streams. For example- HR is having HR core and HR analytics
#and Marketing is having core and Marketing analytics. Analytics is the optional subject and having added extra fees.
#DS is not having analytics. If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.
#If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 . Transportation charges 13000 per semester.
#Calculate the total annual cost based on selected service. User will enter values as subject, analytics(Y/N), Hostel (Y/N),
#food(How many months?), Transportation(semester/annual) Explain the question

core_fee = 200000
analytics_extra = 0.10 * core_fee
hostel_fee = 200000
food_per_month = 2000
transport_per_semester = 13000


subjects = ["HR", "Finance", "Marketing", "DS"]
subject = input("Choose your subject (HR/Finance/Marketing/DS): ")


analytics = input("Do you want analytics? (Y/N): ").upper()


total = core_fee

if analytics == 'Y':
    if subject != 'DS':
        total += analytics_extra
    else:
        print("Analytics not available for DS, only core fee will be charged.")

hostel = input("Do you want hostel? (Y/N): ").upper()
if hostel == 'Y':
    total += hostel_fee

food_months = int(input("How many months of food?: "))
total += food_months * food_per_month


transport = input("Transportation (semester/annual): ").lower()
if transport == 'semester':
    total += transport_per_semester * 2
elif transport == 'annual':
    total += transport_per_semester * 2

print("Total Annual Cost: Rs.", total)
```
