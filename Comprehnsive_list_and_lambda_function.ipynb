```python
#Q.5) Use lambda with filter() to get only odd numbers from a list.
odds=lambda x: x%2!=0
lst=[1,2,3,4]
print(list(filter(odds,lst)))

#OR
list(filter(lambda x: x % 2 != 0, [1, 2, 3, 4]))
```
